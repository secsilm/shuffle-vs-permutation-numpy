```python
import numpy as np
import time
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
from pprint import pprint
%matplotlib inline
```

`res` 是一个字典，key 分别是 `permutation` 和 `shuffle`，测试的时间结果保存在一个元组组成的列表，第一个值是对应数组的长度，第二个值是所用时间。

```python
with open('result.pkl', 'rb') as f:
    res = pickle.load(f)
```

```python
pprint(res)
```

```python
plt.figure(figsize=(15, 8))
plt.plot([x for x, _ in res['permutation']], [y for _, y in res['permutation']])
plt.scatter([x for x, _ in res['permutation']], [y for _, y in res['permutation']])
```

```python
plt.figure(figsize=(13, 7))
plt.plot([y for _, y in res['permutation']])
plt.plot([y for _, y in res['shuffle']])
plt.scatter(np.arange(9), [y for _, y in res['permutation']])
plt.scatter(np.arange(9), [y for _, y in res['shuffle']])
plt.xticks(np.arange(9), ['$10^{}$'.format(i) for i in np.arange(1, 10)], fontsize='large')
plt.xlabel('Length of array')
plt.ylabel('Time/s')
plt.legend(['permutation', 'shuffle'], fontsize='large')
plt.title('shuffle VS permutation', fontsize='xx-large')
plt.tight_layout()
plt.savefig('result.png', dpi=300)
```
